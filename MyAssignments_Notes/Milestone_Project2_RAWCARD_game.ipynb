```python
# This game is a card game [This cell is only for comments]

# We need a class for card
# --> Identify which suit the card belongs to , which card and assign value [SUIT, RANK, VALUE]

# We need a deck class
# --> 52 cards (13 * 4)

# Player class
# Holds cards from decks

# Game logic
# Divide the deck amongst 2 players and check the rank of each card and decide who gets to keep the card
```

```python
# This cell contains all the global variables and imports
# suit and rank can be declared as tuples so that it can be used by all the other classes
# value will be a dictionary which will take in the 'key' and give out the value of the card

import random

suits = ('Hearts','Diamonds','Clubs','Spades')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
value = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10,
         'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
# This cell consists of the card class

class Card:
    
    def __init__(self,suit,rank): # Constructor of the class which requires suit and rank
        self.suit = suit
        self.rank = rank
        self.value = value[rank]
        
    def __str__(self):
        return("This card is the "+ self.suit + " of "+ self.rank)
```

```python
# This cell consists of the deck class
# should shuffle this around (that is why random was imported) -> One method
# should able to deal a card -> Pop one card object from the deck of the card (Another method)
# This class will hold onto a list of objects of cards

class Deck:
    
    def __init__(self): # Default constructor needn't take any inputs
        self.allcards = [] # This will consists of all the card objects taking its size to 52
        
        for suit in suits:
            for rank in ranks:
                Created_Card = Card(suit,rank) # Creates an object of Card
                self.allcards.append(Created_Card) # Add this to the list of cards
    
    def getSize(self):
        return(len(self.allcards))
    
    def shuffleDeck(self):
        random.shuffle(self.allcards) # shuffle is the function from pythons built in library. This fun changes in-place
        
    def dealCard(self):
        return(self.allcards.pop()) # pops the last card and returns it
```

```python
# This cell consists of the player class
# Holds the cards in his hand
# Add/remove cards during the game
# Play --> Should be the top of card (index 0) [Pop[0]]
# Add --> We should add at the bottom of the card the player holds (append method will do)
# Add multiple --> Uses .extend() method which merges the curr list with new list
#                  [Append does not work because it creates nested list]

class Player:
    
    def __init__(self,name):
        self.name = name
        self.allcards = []
        
    def removeCard(self):
        return(self.allcards.pop(0))
    
    def addCards(self,new_cards):
        # There could be multiple card objects or single
        if type(new_cards) == type([]): # new cards has a list of objects
            self.allcards.extend(new_cards)
        else:
            self.allcards.append(new_cards)
        
    def getSize(self):
        return(len(self.allcards))
        
    def __str__(self):
        return(self.name + " has "+str(len(self.allcards)) + " card/s")
```

```python
# This cell consists of the setup and then logic logic of the game
# 1) Create 2 players -> Create a deck and shuffle it -> Once shuffled, divide it between the players
# 2) Check if either one of them have 0 -> If so, the other one has won
# 3) Each player draws their card and we check their value. Lesser value -> Loses his card; 
#                   Higher value -> Gains a card; Equal value [THIS IS RAW!] -> Draw again (These happens in a loop)  
# NOTE: In pt 3. Adding the card needs to go to the bottom (i.e append or extend)

# Created 2 players
player_1 = Player("Jim")
player_2 = Player("Andy")

# Created a deck and shuffled the deck
newDeck = Deck()
newDeck.shuffleDeck()

print(len(newDeck.allcards))

# Run through all cards in deck and give it to each player
for card in range(26):
    # One card dealt to each player in one loop
    player_1.addCards(newDeck.dealCard())
    player_2.addCards(newDeck.dealCard())
    
print(player_1)
print(player_2)
print(len(newDeck.allcards))
```

```python
# Game logic 
# Sample tested --> 226, 221, 270, 650 and even 670 numbers
# Here we are not Shuffling
# When at war, we only choose one card [The longer the game]

ContinueGame = True
# We check how many rounds does this game take to reach a conclusion

numOfRounds = 0

while ContinueGame:
    numOfRounds += 1
    print("Round :"+str(numOfRounds))
    # If either player has 0 cards, that means the other player has won the game
    if (player_1.getSize() == 0 or player_2.getSize() == 0):
        if (player_1.getSize() == 0):
            print("Game has ended "+player_2.name + " has Won!")
            print(player_2)
            print(player_1)
            ContinueGame = False
        else:
            print("Game has ended "+player_1.name + " has Won!")
            print(player_1)
            print(player_2)
            ContinueGame = False
        break
    else:
        # Since both the players have cards, we have to start the game
        # We can also directly add this to a list and compare the [-1] cards from each list
        # listOfPlayer1Cards[-1] > listOfPlayer2Cards[-1] 
        
        player_1Card = player_1.removeCard()
        player_2Card = player_2.removeCard()
        
        if player_1Card.value > player_2Card.value:
            #Player 1 has won this round. Adding back the other and his own card
            player_1.addCards(player_1Card)
            player_1.addCards(player_2Card)
        elif player_1Card.value < player_2Card.value:
            # Player 2 has won this round. Adding back the other and his own card
            player_2.addCards(player_2Card)
            player_2.addCards(player_1Card)
        else:
            # We are at WAR!!. Therefore we continue to play each hand
            listofPlayer1Cards = []
            listofPlayer1Cards.append(player_1Card)
            listofPlayer2Cards = []
            listofPlayer2Cards.append(player_2Card)
            
            atWar = True
            while atWar:
                # Remove cards from each player and again check their values
                atWarPlayer_1Card = player_1.removeCard()
                atWarPlayer_2Card = player_2.removeCard()
                print("WAR!!")
                if atWarPlayer_1Card.value > atWarPlayer_2Card.value:
                    # Player one has won this round
                    listofPlayer2Cards.append(atWarPlayer_2Card)
                    listofPlayer1Cards.append(atWarPlayer_1Card)
                    player_1.addCards(listofPlayer1Cards)
                    player_1.addCards(listofPlayer2Cards)
                    atWar = False
                    break
                elif atWarPlayer_1Card.value < atWarPlayer_2Card.value:
                    # Player two has won this round
                    listofPlayer2Cards.append(atWarPlayer_2Card)
                    listofPlayer1Cards.append(atWarPlayer_1Card)
                    player_2.addCards(listofPlayer2Cards)
                    player_2.addCards(listofPlayer1Cards)
                    atWar = False
                    break
                else:
                    # War is still ongoing
                    listofPlayer2Cards.append(atWarPlayer_2Card)
                    listofPlayer1Cards.append(atWarPlayer_1Card)
                    continue
```
