```python
# This is first milestone project to come up with a tic-tac-toe game
# It needs to accept user input and complete the board accordingly

# Display the board game first (3*3 matrix)

def displayboard(row1,row2,row3):
    print(row1)
    #print('     |     |    ')
    print(' ---------------')
    print(row2)
    #print('     |     |    ')
    print(' ---------------')
    print(row3)
    #print('     |     |    ')
    
def convert_user_to_tableindex(row1,row2,row3,userinputindex,user):
    # This function uses string slicing
    if userinputindex[0] == 1:
        #Only inputs for row1
        if userinputindex[1] == 1:
            row1 = row1[:3]+ row1[3:4].replace(' ',user) + row1[4:]
            # Row1[3] consists userentry
        elif userinputindex[1] == 2:
            row1 = row1[:8]+ row1[8:9].replace(' ',user) + row1[9:]
        elif userinputindex[1] == 3:
            row1 = row1[:13]+ row1[13:14].replace(' ',user) + row1[14:]
        else:
            pass
    elif userinputindex[0] == 2:
        #Only inputs for row2
        if userinputindex[1] == 1:
            row2 = row2[:3]+ row2[3:4].replace(' ',user) + row2[4:]
        elif userinputindex[1] == 2:
            row2 = row2[:8]+ row2[8:9].replace(' ',user) + row2[9:]
        elif userinputindex[1] == 3:
            row2 = row2[:13]+ row2[13:14].replace(' ',user) + row2[14:]
        else:
            pass
    elif userinputindex[0] == 3:
        #Only inputs for row3
        if userinputindex[1] == 1:
            row3 = row3[:3]+ row3[3:4].replace(' ',user) + row3[4:]
        elif userinputindex[1] == 2:
            row3 = row3[:8]+ row3[8:9].replace(' ',user) + row3[9:]
        elif userinputindex[1] == 3:
            row3 = row3[:13]+ row3[13:14].replace(' ',user) + row3[14:]
        else:
            pass
    else:
        pass
    return(row1,row2,row3)

def did_someone_win(row1,row2,row3,firstuser,seconduser):
    
    if ((row1[3] == firstuser and row1[8] == firstuser and row1[13] == firstuser) or
        (row2[3] == firstuser and row2[8] == firstuser and row2[13] == firstuser) or
        (row3[3] == firstuser and row3[8] == firstuser and row3[13] == firstuser) or
        (row1[3] == firstuser and row2[3] == firstuser and row3[3] == firstuser) or
        (row1[8] == firstuser and row2[8] == firstuser and row3[8] == firstuser) or
        (row1[13] == firstuser and row2[13] == firstuser and row3[13] == firstuser) or
        (row1[3] == firstuser and row2[8] == firstuser and row3[13] == firstuser) or
        (row1[13] == firstuser and row2[8] == firstuser and row3[3] == firstuser)):
        print('Victory!! for '+firstuser)
        return True
    elif ((row1[3] == seconduser and row1[8] == seconduser and row1[13] == seconduser) or
        (row2[3] == seconduser and row2[8] == seconduser and row2[13] == seconduser) or
        (row3[3] == seconduser and row3[8] == seconduser and row3[13] == seconduser) or
        (row1[3] == seconduser and row2[3] == seconduser and row3[3] == seconduser) or
        (row1[8] == seconduser and row2[8] == seconduser and row3[8] == seconduser) or
        (row1[13] == seconduser and row2[13] == seconduser and row3[13] == seconduser) or
        (row1[3] == seconduser and row2[8] == seconduser and row3[13] == seconduser) or
        (row1[13] == seconduser and row2[8] == seconduser and row3[3] == seconduser)):
        print('Victory!! for '+seconduser)
        return True
    else:
        return False
```

```python
# Function to check if the first user wants to be X or O
def chooseXorO():
    choice = input('Choose between 0 and 1 ==> 0 for O and 1 for X : ' )
    if choice == '1' or choice == '0':
        return choice
    else:
        print('Incorrect entry!! Start again')
        return False
    
def printuserchoice(firstuser,seconduser):
    print('The first user has chosen -> '+firstuser)
    print('The second user has by default been assigned -> '+seconduser)
```

```python
# Function to input the row and then the exact index where he/she wants their symbol to be inserted
# Keep asking if the input is wrong
def getindexforGame(user):
    is_row_valid = False
    is_index_valid = False
    correctnumbers = [1,2,3]
    final_input_row = 0
    final_input_index = 0
    while not is_index_valid or not is_row_valid:
        input_row = input('Enter the row the '+user+' wants to place his mark on: ')
        try:
            final_input_row = int(input_row)
            if final_input_row in correctnumbers:
                is_row_valid = True
                input_index = input('Enter the index the '+user+' wants to place his mark on: ')
                try:
                    final_input_index = int(input_index)
                    if final_input_index in correctnumbers:
                        is_index_valid = True
                        return(final_input_row, final_input_index)
                except ValueError:
                    print('Incorrect entry!! Start again')
            else:
                print('Incorrect entry!! Start entry again')
        except ValueError:
            print('Incorrect entry!! Start entry again')
            
# Define a function to validate the entry chosen is unique or not
def validate_userEntry(user,userinputindex,assignedListindexes):
        uniqueentry = False
        if userinputindex in assignedListindexes:
            print('This entry has been chosen already, no overwriting possible')
            while not uniqueentry:
                userinputindex = getindexforGame(user)
                if userinputindex not in assignedListindexes:
                    assignedListindexes.append(userinputindex)
                    uniqueentry = True
                else:
                    pass
        else:
            assignedListindexes.append(userinputindex)
```

```python
# Let's define the main function here
def main_function():
    # For the timebeing assing each row to a string
    row1 = '     |     |    '
    row2 = '     |     |    '
    row3 = '     |     |    '
    # Print the board 
    displayboard(row1,row2,row3)
    print(' ')
    print(' ')
    print('                     LET THE GAMES BEGIN!!!')
    
    # Ask the user what's his preferred choice
    validchoice = False
    firstuser = ''
    seconduser = ''
    while not validchoice:
        validchoice = chooseXorO()
        if validchoice=='1':
            firstuser = 'X'
            seconduser = 'O'
            validchoice = True
        elif validchoice == '0':
            firstuser = 'O'
            seconduser = 'X'
            validchoice = True
        else:
            pass
    
    # Print the users choice just in case
    printuserchoice(firstuser,seconduser)
    
    # Ask inputs from the first user, show the board and then ask input from second user
    assignedListindexes = []
    userinputindex = (0,0)
    print(' ')
    print(' ')
    draw = False
    while True:
        if did_someone_win(row1,row2,row3,firstuser,seconduser) or draw:
            break
        print(firstuser + ' : Your move')
        userinputindex = getindexforGame(firstuser)
        validate_userEntry(firstuser,userinputindex,assignedListindexes)
        (row1,row2,row3) = convert_user_to_tableindex(row1,row2,row3,userinputindex,firstuser)
        displayboard(row1,row2,row3)
        if did_someone_win(row1,row2,row3,firstuser,seconduser) or draw:
            break
        print(seconduser + ' :Your move')
        userinputindex = getindexforGame(seconduser)
        validate_userEntry(seconduser,userinputindex,assignedListindexes)
        (row1,row2,row3) = convert_user_to_tableindex(row1,row2,row3,userinputindex,seconduser)
        displayboard(row1,row2,row3)
        if (len(assignedListindexes) > 8):
            draw = True
            print('Match ended in a draw')            
```
